```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
try:
    df = pd.read_csv('/kaggle/input/creditcardfraud/creditcard.csv')  # Kaggle
except FileNotFoundError:
    try:
        df = pd.read_csv('/content/drive/MyDrive/Data sets/credit card ml project/creditcard.csv')    # Google Drive (Colab)
    except FileNotFoundError:
        df = pd.read_csv("E:/AI ML/Project/Data/creditcard.csv")                           # Local
```

```python
# imports
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
```

```python
# defining x and y
x = df.drop('Class', axis=1)
y = df['Class']
```

```python
# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42, stratify=y)
```

```python
# scaling the data
scaler = RobustScaler()
x_train_robust = scaler.fit_transform(x_train)
x_test_robust = scaler.transform(x_test)
```

```python
# saving the x and y data
import os, joblib

save_dir = "/content/drive/MyDrive/AI ML/Project/Preprocessed/artifacts"
os.makedirs(save_dir, exist_ok=True)

save_path = os.path.join(save_dir, "preprocessed_x_y_data.pkl")

joblib.dump((x_train_robust, y_train, x_test_robust, y_test), save_path)

print("Saved to:", save_path)
print("Files in artifacts:", os.listdir(save_dir))
```
